**The FinanceToolkit** is an open-source toolkit in which all relevant financial ratios (100+), indicators and performance measurements are written down in the most simplistic way allowing for complete transparency of the calculation method. This allows you to not have to rely on metrics from other providers and, given a financial statement, allow for efficient manual calculations. This leads to one uniform method of calculation being applied that is available and understood by everyone.

````python
import pandas as pd
import matplotlib.pyplot as plt

from financetoolkit import Toolkit

API_KEY = "FINANCIAL_MODELING_PREP_API_KEY"
````

**Initializing only is required once.** This is the case for any function so once you have obtained a balance sheet statement, it will be stored accordingly which means that requests to FinancialModelingPrep, the source used in these examples, are kept to a minimum. Note that in this example annual data is used but by adding `quarterly=True` to the Toolkit initialization, quarterly data can also be collected. Note that this requires a Premium subscription from FMP.

````python
# Initialize the Toolkit with company tickers
companies = Toolkit(
    ["AAPL", "MSFT", "GOOGL", "AMZN"], api_key=API_KEY, start_date="2005-01-01"
)
````

The Options module is meant to calculate important options metrics such as the First, Second and Third Order Greeks, the Black Scholes Model and the Option Chains as well as Implied Volatilities, Breeden—Litzenberger and more. A great start is the Black Scholes model which calculates the theoretical value of an option.

````python
companies.options.get_black_scholes_model()
````

It is also possible to calculate the theoretical values based on the Binomial Model, this illustrates the option valuation diagram. The Binomial Model can determine both American and European options in which the crucial difference is that the American option can be exercised at any time before the expiration date, while the European option can only be exercised at the expiration date.

````python
companies.options.get_binomial_model(american_option=True)
````

Underlying the Binomial Model is the up and down movement of the stock price. The up and down movement is calculated using the volatility of the stock and the time to expiration. It is possible to show these movements in a diagram which visually depicts the Binomial Tree.

````python
stock_price_simulations = companies.options.get_stock_price_simulation(timesteps=10)

fig, ax = plt.subplots(2, 2, figsize=(15, 10))

stock_price_simulations_transposed = stock_price_simulations.T
stock_price_simulations_transposed.index = (
    stock_price_simulations_transposed.index.astype("datetime64[ns]")
)

for i, ticker in enumerate(companies._tickers):
    ax[i // 2, i % 2].plot(stock_price_simulations_transposed[ticker])
    ax[i // 2, i % 2].set_title(ticker)
    ax[i // 2, i % 2].xaxis.set_tick_params(rotation=20)
    ax[i // 2, i % 2].yaxis.set_tick_params(labelsize=8)
    ax[i // 2, i % 2].grid(linestyle="--", alpha=0.5)

fig.suptitle("Stock Price Simulations", fontweight="bold")
fig.tight_layout()
plt.show()
````

Based on these theoretical values and the input that is collected throughout the Toolkit (the historical data, risk free rate, dividend yield and more) it is possible to calculate any type of Greek. E.g. a simple example is the Delta which is the change in the option price given a change in the underlying asset price.

````python
companies.options.get_delta()
````

The parameters can be altered to lengthen or shorten the Strike Price and Time to Expiration range as well as the date for each ticker's stock price. By default, the most recent price is used, the Strike Prices 75% and 125% of the price are used and the time to expiration is set to the next 30 days. All of this can be altered e.g. for Gamma. Here, it helps to set `show_input_info=True` to understand the underlying parameters.

````python
companies.options.get_gamma(
    start_date="2020-01-02",
    strike_price_range=0.2,
    expiration_time_range=180,
    show_input_info=True,
)
````

Instead of one option, it is possible to plot all First, Second and Third Order Greeks. This can be done with the `collect_` functions in which you can choose whether to plot all or one of the orders. E.g. all Greeks can be plotted with the following.

````python
companies.options.collect_all_greeks()
````

Perhaps you are looking to fluctuate with the Stock Price instead of the Strike Price. This is where the architecture of the Finance Toolkit really shines given that all equations are written down seperately. For example, let's plot the Delta and Gamma for a range of Stock Prices.

````python
from financetoolkit.options import greeks_model

delta_values_call = {}
delta_values_put = {}
gamma_values = {}

STRIKE_PRICE = 100
RISK_FREE_RATE = 0.05
TIME_TO_EXPIRATION = 180 / 365
VOLATILITY = 0.2
DIVIDEND_YIELD = 0.01

for stock_price in range(50, 150, 1):
    delta_values_call[stock_price] = greeks_model.get_delta(
        stock_price=stock_price,
        strike_price=STRIKE_PRICE,
        risk_free_rate=RISK_FREE_RATE,
        time_to_expiration=TIME_TO_EXPIRATION,
        volatility=VOLATILITY,
        dividend_yield=DIVIDEND_YIELD,
    )

    delta_values_put[stock_price] = greeks_model.get_delta(
        stock_price=stock_price,
        strike_price=STRIKE_PRICE,
        risk_free_rate=RISK_FREE_RATE,
        time_to_expiration=TIME_TO_EXPIRATION,
        volatility=VOLATILITY,
        dividend_yield=DIVIDEND_YIELD,
        put_option=True,
    )

    gamma_values[stock_price] = greeks_model.get_gamma(
        stock_price=stock_price,
        strike_price=STRIKE_PRICE,
        risk_free_rate=RISK_FREE_RATE,
        time_to_expiration=TIME_TO_EXPIRATION,
        volatility=VOLATILITY,
        dividend_yield=DIVIDEND_YIELD,
    )


delta_values_call_df = pd.Series(delta_values_call)
delta_values_put_df = pd.Series(delta_values_put)
gamma_values_df = pd.Series(gamma_values)

delta_values_call_df.name, delta_values_put_df.name, gamma_values_df.name = (
    "$100 Call Delta",
    "$100 Put Delta",
    "$100 Gamma",
)

delta_values_call_df.plot(ylabel="Delta", figsize=(15, 3), color="red", legend=True)
delta_values_put_df.plot(ylabel="Delta", figsize=(15, 3), color="blue", legend=True)
gamma_values_df.plot(
    title="Gamma and Delta vs Stock Price",
    xlabel="Stock Price",
    ylabel="Gamma",
    color="green",
    secondary_y=True,
    figsize=(15, 3),
    legend=True,
)
````
